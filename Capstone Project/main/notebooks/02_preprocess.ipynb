```python
# ============================================================
# 1. Import Libraries
# ============================================================

import sys
from pathlib import Path

import pandas as pd

# Add project root to Python path
sys.path.append(str(Path().resolve().parent))
```

```python
# ============================================================
# 2. Import Custom Modules
# ============================================================

from src.data_loader import load_data

from src.preprocess import (
    encode_target,
    drop_id_column,
    split_features_target
)
```

```python
# ============================================================
# 3. Load Dataset
# ============================================================

df = load_data()

# Preview dataset
df.head()
```

```python
# ============================================================
# 4. Encode Target Variable
# ============================================================

df = encode_target(df)

print("Encoded Target Distribution:")

print(df['Diagnosis'].value_counts())
```

```python
# ============================================================
# 5. Remove ID Column
# ============================================================

df = drop_id_column(df)

print("Updated Dataset Shape:")

print(df.shape)
```

```python
# ============================================================
# 6. Split Features and Target
# ============================================================

X, y = split_features_target(df)

print("Feature Matrix Shape:", X.shape)
print("Target Vector Shape:", y.shape)
```

```python
# ============================================================
# 7. Create Final Processed Dataset
# ============================================================

processed_df = X.copy()

processed_df['Diagnosis'] = y

processed_df.head()
```

```python
# ============================================================
# 8. Save Processed Dataset
# ============================================================

processed_path = Path("../data/processed/processed_data.csv")

processed_df.to_csv(
    processed_path,
    index=False
)

print(f"Processed dataset saved at: {processed_path}")
```

# Preprocessing Summary

## Transformations Performed

1. The target column `Diagnosis` was encoded as:
   - Malignant (M) → 1
   - Benign (B) → 0

2. The `ID` column was removed because it does not contribute meaningful predictive information.

3. The dataset was separated into:
   - Feature matrix (`X`)
   - Target vector (`y`)

4. The final processed dataset was saved for downstream modeling and evaluation tasks.

## Why These Steps Were Necessary

- Machine learning models require numerical target labels.
- Identifier columns can introduce noise and potential overfitting.
- Separating features and target improves modularity and pipeline design.
- Saving processed data improves reproducibility and avoids repeated preprocessing.
