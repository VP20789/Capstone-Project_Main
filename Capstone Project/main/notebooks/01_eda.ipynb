```python
# ============================================================
# 1. Import Libraries
# ============================================================

import sys
from pathlib import Path

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Add project root to Python path
sys.path.append(str(Path().resolve().parent))

# Visualization settings
sns.set_style("whitegrid")
pd.set_option("display.max_columns", None)
```

```python
# ============================================================
# 2. Import Custom Modules
# ============================================================

from src.data_loader import load_data

from src.validate import (
    dataset_shape,
    missing_values,
    duplicate_rows,
    diagnosis_distribution,
    feature_count
)
```

```python
# ============================================================
# 3. Load Dataset
# ============================================================

df = load_data()

# Display first 5 rows
df.head()
```

```python
# ============================================================
# 4. Dataset Overview
# ============================================================

print("Dataset Shape:", dataset_shape(df))
print("Feature Count:", feature_count(df))
```

```python
# ============================================================
# 5. Missing Values Check
# ============================================================

print("Total Missing Values:", missing_values(df))
```

```python
# ============================================================
# 6. Duplicate Check
# ============================================================

print("Duplicate Rows Before Removal:", duplicate_rows(df))

# Remove duplicates if any
df = df.drop_duplicates()

print("Duplicate Rows After Removal:", duplicate_rows(df))
```

```python
# ============================================================
# 7. Dataset Information
# ============================================================

df.info()
```

```python
# ============================================================
# 8. Statistical Summary
# ============================================================

df.describe()
```

```python
# ============================================================
# 9. Target Variable Distribution
# ============================================================

diagnosis_counts = diagnosis_distribution(df)

print(diagnosis_counts)

# Plot diagnosis distribution
plt.figure(figsize=(6, 4))

sns.countplot(
    x='Diagnosis',
    data=df
)

plt.title("Diagnosis Class Distribution")
plt.xlabel("Diagnosis")
plt.ylabel("Count")

plt.show()
```

```python
# ============================================================
# 10. Percentage Distribution
# ============================================================

diagnosis_percent = (
    df['Diagnosis']
    .value_counts(normalize=True) * 100
)

print(diagnosis_percent)
```

```python
# ============================================================
# 11. Encode Target Variable
# M = 1 (Malignant)
# B = 0 (Benign)
# ============================================================

df['Diagnosis'] = df['Diagnosis'].map({
    'M': 1,
    'B': 0
})

# Verify encoding
df['Diagnosis'].value_counts()
```

```python
# ============================================================
# 12. Drop ID Column
# ============================================================

df.drop(columns=['ID'], inplace=True)

print("Updated Shape:", df.shape)
```

```python
# ============================================================
# 13. Correlation Matrix
# ============================================================

plt.figure(figsize=(20, 16))

correlation_matrix = df.corr()

sns.heatmap(
    correlation_matrix,
    cmap='coolwarm',
    linewidths=0.5
)

plt.title("Feature Correlation Matrix")

plt.show()
```

```python
# ============================================================
# 14. Distribution of Key Features
# ============================================================

key_features = [
    'radius_mean',
    'texture_mean',
    'perimeter_mean',
    'area_mean',
    'smoothness_mean'
]

for feature in key_features:

    plt.figure(figsize=(6, 4))

    sns.histplot(
        data=df,
        x=feature,
        kde=True
    )

    plt.title(f"Distribution of {feature}")

    plt.show()
```

```python
# ============================================================
# 15. Boxplots for Outlier Detection
# ============================================================

for feature in key_features:

    plt.figure(figsize=(6, 4))

    sns.boxplot(
        x=df[feature]
    )

    plt.title(f"Boxplot of {feature}")

    plt.show()
```

```python
# ============================================================
# 16. Pairplot (Optional - Small Feature Set)
# ============================================================

selected_features = [
    'Diagnosis',
    'radius_mean',
    'texture_mean',
    'perimeter_mean',
    'area_mean'
]

sns.pairplot(
    df[selected_features],
    hue='Diagnosis'
)

plt.show()
```

```python
# ============================================================
# 17. Save Cleaned Dataset
# ============================================================

processed_path = Path("../data/interim/breast_cancer_cleaned.csv")

df.to_csv(
    processed_path,
    index=False
)

print(f"Cleaned dataset saved to: {processed_path}")
```
