```python
# ============================================================
# 1. Import Libraries
# ============================================================

import sys
from pathlib import Path

import pandas as pd
import joblib

# Add project root to Python path
sys.path.append(str(Path().resolve().parent))
```

```python
# ============================================================
# 2. Import Custom Modules
# ============================================================

from src.split import stratified_split

from src.train import (
    create_random_forest_model,
    train_model,
    generate_predictions
)
```

```python
# ============================================================
# 3. Load Processed Dataset
# ============================================================

df = pd.read_csv(
    "../data/processed/processed_data.csv"
)

df.head()
```

```python
# ============================================================
# 4. Split Features and Target
# ============================================================

X = df.drop(columns=['Diagnosis'])

y = df['Diagnosis']

print("Feature Matrix Shape:", X.shape)
print("Target Vector Shape:", y.shape)
```

```python
# ============================================================
# 5. Perform Stratified Split
# ============================================================

(
    X_train,
    X_val,
    X_test,
    y_train,
    y_val,
    y_test
) = stratified_split(X, y)
```

```python
# ============================================================
# 6. Print Split Shapes
# ============================================================

print("Training Set Shape:", X_train.shape)
print("Validation Set Shape:", X_val.shape)
print("Test Set Shape:", X_test.shape)
```

```python
# ============================================================
# 7. Check Class Distribution
# ============================================================

print("\nTraining Distribution:")
print(y_train.value_counts(normalize=True))

print("\nValidation Distribution:")
print(y_val.value_counts(normalize=True))

print("\nTest Distribution:")
print(y_test.value_counts(normalize=True))
```

```python
# ============================================================
# 8. Create Random Forest Model
# ============================================================

model = create_random_forest_model()

print(model)
```

```python
# ============================================================
# 9. Train Model
# ============================================================

model = train_model(
    model,
    X_train,
    y_train
)

print("Model training completed.")
```

```python
# ============================================================
# 10. Generate Validation Predictions
# ============================================================

val_predictions, val_probabilities = generate_predictions(
    model,
    X_val
)

print("Validation predictions generated.")
```

```python
# ============================================================
# 11. Generate Test Predictions
# ============================================================

test_predictions, test_probabilities = generate_predictions(
    model,
    X_test
)

print("Test predictions generated.")
```

```python
# ============================================================
# 12. Save Trained Model
# ============================================================

model_path = "../artifacts/models/rf_model.joblib"

joblib.dump(
    model,
    model_path
)

print(f"Model saved at: {model_path}")
```

# Training Summary

## Model Training Workflow

1. The processed dataset was loaded successfully.

2. The dataset was split into:
   - Training Set (70%)
   - Validation Set (15%)
   - Test Set (15%)

3. Stratified splitting was used to preserve class distribution across all datasets.

4. A Random Forest Classifier was initialized using a fixed random state for reproducibility.

5. The model was trained using the training dataset.

6. Predictions and probability scores were generated for:
   - Validation dataset
   - Test dataset

7. The trained model was saved for future evaluation and inference tasks.

## Why These Steps Matter

- Train/Validation/Test separation prevents data leakage.
- Stratification ensures balanced class representation.
- Validation data supports unbiased tuning and threshold selection.
- Probability scores are required for ROC-AUC and threshold analysis.
- Saving the trained model improves reproducibility and deployment readiness.
